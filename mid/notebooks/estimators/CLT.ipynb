```
from scipy import stats
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```
np.random.randint(1, 7, 10)
```

```
def sample_dice_roll(n_samples):
    return np.random.randint(1, 7, n_samples)
```

```
experiments = np.array([np.mean(sample_dice_roll(10)) for _ in range(10_000)])
```

```
scales = (5, 10, 50, 100, 1000, 10000)
colors = sns.diverging_palette(10, 220, sep=80, n=7)
```

```
figure, axes = plt.subplots(figsize=[20, 12])

for exponent, (scale, color) in enumerate(zip(scales, colors)):
    sns.kdeplot(
        experiments[:scale],
        color=color,
        shade=False,
        label=scale
        )

plt.legend(prop={'size': 20})
axes.grid()
```

```
plt.figure(figsize=[20, 12])

for scale, color in zip(scales, colors):
    sns.distplot(experiments[:scale], color=color, kde=True, norm_hist=True)

plt.grid()
```
