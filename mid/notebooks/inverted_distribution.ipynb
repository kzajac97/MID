```python
t = np.linspace(0, 5, 10_000)
```

```python
plt.figure(figsize=[12, 8])

plt.plot(t, np.array([triangle(point) for point in t]))
plt.plot(t, np.array([exponential(point, 1.0) for point in t]))
plt.plot(t, cauchy(t, x0=0.0, gamma=1.0))
plt.plot(t, laplace(t, mu=1.0, b=1.0))
plt.plot(t, logistic(t, mu=1.0, s=1.0))

plt.legend(['Triangle', 'Exp', 'Cauchy', 'Laplace', 'Logistic'], prop={'size': 20})

plt.grid()
```

```python
plt.figure(figsize=[12, 8])

# plt.plot(t, np.array([cdf(triangle, point) for point in t]))
plt.plot(t, np.array([cdf(exponential, point, 1.0) for point in t]))
plt.plot(t, np.array([cdf(cauchy, point, 0.0, 1.0) for point in t]))
plt.plot(t, np.array([cdf(laplace, point, 1.0, 1.0) for point in t]))
plt.plot(t, np.array([cdf(logistic, point, 1.0, 1.0) for point in t]))

plt.legend(['Exp', 'Cauchy', 'Laplace', 'Logistic'], prop={'size': 20})

plt.grid()
```
