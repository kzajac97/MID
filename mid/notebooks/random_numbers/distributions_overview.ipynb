### Modules imports

```python
%autoreload 2

from source.functions.distributions import cauchy, exponential, laplace, logistic, triangle
from source.functions.cumulative import cumulative_cauchy, cumulative_exponential, cumulative_laplace, cumulative_logistic, cumulative_triangle
from source.generators.inverted_distribution import cumulative_distribution_function
```

```python
t = np.linspace(-3, 3, 10_000)
```

```python
plt.figure(figsize=[20, 8])

plt.plot(t, np.array([triangle(point) for point in t]))
plt.plot(t, np.array([exponential(point) for point in t]))
plt.plot(t, cauchy(t))
plt.plot(t, laplace(t))
plt.plot(t, logistic(t))

plt.legend(['Triangle', 'Exp', 'Cauchy', 'Laplace', 'Logistic'], prop={'size': 20})

plt.grid()

plt.savefig(r'../images/distributions/overview/pdf.png')
```

```python
plt.figure(figsize=[20, 8])

plt.plot(t, np.array([cumulative_triangle(point) for point in t]))
plt.plot(t, np.array([cumulative_exponential(point) for point in t]))
plt.plot(t, cumulative_cauchy(t))
plt.plot(t, np.array([cumulative_laplace(point) for point in t]))
plt.plot(t, cumulative_logistic(t))

plt.legend(['Triangle', 'Exp', 'Cauchy', 'Laplace', 'Logistic'], prop={'size': 20})

plt.grid()

plt.savefig(r'../images/distributions/overview/analytical_cdf.png')
```

```python
plt.figure(figsize=[20, 8])

plt.plot(t, np.array([cumulative_distribution_function(triangle, point) for point in t]))
plt.plot(t, np.array([cumulative_distribution_function(exponential, point) for point in t]))
plt.plot(t, np.array([cumulative_distribution_function(cauchy, point) for point in t]))
plt.plot(t, np.array([cumulative_distribution_function(laplace, point) for point in t]))
plt.plot(t, np.array([cumulative_distribution_function(logistic, point, infinity_approximation=100) for point in t]))

plt.legend(['Triangle', 'Exp', 'Cauchy', 'Laplace', 'Logistic'], prop={'size': 20})

plt.grid()

plt.savefig(r'../images/distributions/overview/numerical_cdf.png')
```
